```python
import numpy as np
import json

# MATRICES (las que ya tienes)
rf_cm = np.array([[2425, 2422],
                  [1720, 2612]])

xgb_cm = np.array([[2532, 2315],
                   [1945, 2387]])

np.save(f"{BASE_PATH}/rf_confusion.npy", rf_cm)
np.save(f"{BASE_PATH}/xgb_confusion.npy", xgb_cm)

metrics = {
    "rf": {
        "accuracy": 0.55,
        "roc_auc": 0.583,
        "f1": 0.55
    },
    "xgb": {
        "accuracy": 0.54,
        "roc_auc": 0.548,
        "f1": 0.54
    }
}

with open(f"{BASE_PATH}/metrics.json", "w") as f:
    json.dump(metrics, f, indent=4)
```
